```python
import random

def generate_test_file(filename, count):
    """Genera un archivo con N números aleatorios y algunos errores."""
    with open(filename, 'w', encoding='utf-8') as f:
        for _ in range(count):
            f.write(f"{random.uniform(1, 1000):.2f}\n")
        # Agregamos un par de datos inválidos para probar el Req 3
        f.write("DATO_INVALIDO\n")
        f.write("99.99.99\n")

if __name__ == "__main__":
    generate_test_file("data_exercise_1.txt", 1000)
    print("Archivo 'data_exercise_1.txt' generado con 1000 registros.")
```
